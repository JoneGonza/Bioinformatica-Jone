1.- Abre el fichero dna.txt, guarda una lista del contenido del archivo. Devuelve su longitud y su primer caracter

```python
# Abrir el archivo "dna.txt" en modo lectura ("r")
# 'with open(...)' garantiza que el archivo se cierre autom√°ticamente al terminar
with open("dna.txt", "r") as f:
    # Leemos todo el contenido del archivo como una sola cadena de texto
    contenido = f.read().strip()  
    # .strip() elimina los espacios en blanco o saltos de l√≠nea al inicio y al final

# Convertimos la cadena en una lista de caracteres individuales
# Por ejemplo, "ATGC" ‚Üí ['A', 'T', 'G', 'C']
lista_dna = list(contenido)

# Calculamos la longitud total de la secuencia de ADN (n√∫mero de caracteres)
longitud = len(lista_dna)

# Guardamos el primer car√°cter de la secuencia (posici√≥n 0 en la lista)
primer_caracter = lista_dna[0]

# Mostramos los resultados por pantalla
print("Longitud de la secuencia:", longitud)
print("Primer car√°cter:", primer_caracter)
```

2.Crea dos archivos. Uno almacena los accession que empiezen por 'a' y el otro el resto

```python
# Lista de accessions que vamos a clasificar
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']

# Abrimos dos archivos al mismo tiempo:
#  - "empiezan_a.txt" ‚Üí para los que empiezan por 'a'
#  - "resto.txt" ‚Üí para todos los dem√°s
with open("empiezan_a.txt", "w") as f_a, open("resto.txt", "w") as f_resto:
    for acc in accs:
        # startswith('a') comprueba si el texto empieza por la letra 'a'
        if acc.startswith('a'):
            # Si empieza por 'a', lo escribimos en el archivo 'empiezan_a.txt'
            f_a.write(acc + "\n")
        else:
            # Si no empieza por 'a', lo escribimos en 'resto.txt'
            f_resto.write(acc + "\n")

# Cuando termina el bloque 'with', los archivos se cierran autom√°ticamente
print("Archivos creados correctamente:")
print("- empiezan_a.txt (accesiones que empiezan por 'a')")
print("- resto.txt (las dem√°s)")

```

##### 4.- Escribe una funci√≥n que cuente el porcentaje de A de una secuencia

```python
def porcentaje_A(secuencia):
    # Convertimos toda la secuencia a may√∫sculas
    # Esto garantiza que 'a' y 'A' se traten igual
    secuencia = secuencia.upper()

    # Calculamos la longitud total de la secuencia
    total = len(secuencia)

    # Si la secuencia est√° vac√≠a (longitud = 0),
    # devolvemos 0 para evitar un error al dividir por cero
    if total == 0:
        return 0  # evita divisi√≥n por cero

    # Contamos cu√°ntas 'A' hay en la secuencia
    num_A = secuencia.count('A')

    # Calculamos el porcentaje:
    # (n√∫mero de A / longitud total) * 100
    porcentaje = (num_A / total) * 100

    # Devolvemos el porcentaje calculado
    return porcentaje


# ==== Ejemplo de uso ====

# Definimos una secuencia de ADN
seq = "ATGCAAACTGA"

# Llamamos a la funci√≥n y mostramos el resultado
print("Porcentaje de A:", porcentaje_A(seq), "%")
# üëâ Salida esperada: Porcentaje de A: 36.36363636363637 %
```

##### ¬øPuedes hacer que devuelva dos decimales? ¬øY que lidie con may√∫sculas y min√∫sculas?

```python
def porcentaje_A(secuencia):
    # Definimos una funci√≥n llamada 'porcentaje_A' que calcula
    # el porcentaje de la base 'A' (adenina) en una secuencia de ADN.

    # Convertimos toda la secuencia a may√∫sculas para que
    # las 'a' y las 'A' se traten igual.
    secuencia = secuencia.upper()

    # Calculamos la longitud total de la secuencia.
    total = len(secuencia)

    # Si la secuencia est√° vac√≠a (longitud 0),
    # devolvemos 0.0 para evitar un error de divisi√≥n por cero.
    if total == 0:
        return 0.0  # evita divisi√≥n por cero

    # Contamos cu√°ntas veces aparece la letra 'A' en la secuencia.
    num_A = secuencia.count('A')

    # Calculamos el porcentaje de 'A' en la secuencia:
    # (n√∫mero de A / total de bases) * 100
    porcentaje = (num_A / total) * 100

    # Redondeamos el resultado a 2 decimales para mostrarlo m√°s claro.
    return round(porcentaje, 2)


# ==== Ejemplo de uso ====

# Definimos una secuencia de ADN con letras may√∫sculas y min√∫sculas.
seq = "aTgCAaAcTGa"

# Llamamos a la funci√≥n y mostramos el resultado por pantalla.
print("Porcentaje de A:", porcentaje_A(seq), "%")
# üëâ Salida esperada: Porcentaje de A: 36.36 %

```

¬øY que devuelva el n√∫meo de decimales que yo quiero?

```python
def porcentaje_A(secuencia, decimales=2):
    # Definimos una funci√≥n llamada 'porcentaje_A' que calcula
    # el porcentaje de adeninas (A) en una secuencia de ADN.
    # Tiene dos par√°metros:
    #  - secuencia ‚Üí la cadena de ADN
    #  - decimales ‚Üí n√∫mero de decimales que queremos (por defecto = 2)

    # Convertimos toda la secuencia a may√∫sculas
    # Esto permite contar correctamente las 'A' aunque haya min√∫sculas
    secuencia = secuencia.upper()

    # Calculamos la longitud total de la secuencia
    total = len(secuencia)

    # Si la secuencia est√° vac√≠a, devolvemos 0.0 para evitar un error al dividir
    if total == 0:
        return 0.0  # evita divisi√≥n por cero

    # Contamos cu√°ntas 'A' hay en la secuencia
    num_A = secuencia.count('A')

    # Calculamos el porcentaje: (n√∫mero de A / total) * 100
    porcentaje = (num_A / total) * 100

    # Redondeamos el resultado al n√∫mero de decimales indicado
    return round(porcentaje, decimales)


# ===== Ejemplo de uso =====

seq = "aTgCAaAcTGa"  # Secuencia de ADN con may√∫sculas y min√∫sculas

# Llamamos a la funci√≥n con el valor por defecto (2 decimales)
print("Porcentaje de A (2 decimales):", porcentaje_A(seq))

# Llamamos a la funci√≥n indicando 4 decimales
print("Porcentaje de A (4 decimales):", porcentaje_A(seq, 4))

```

#### 5.- De la lista de accessions que se proporciona obt√©n lo siguiente:
1. Los que tengan el n√∫mero 5
2. Los que tengan la letra 'd' o la 'e'
3. Los que tengan las letras 'd' y 'e' en ese orden
4. Los que tengan las letras 'd' y 'e' en ese orden con una letra cualquiera en medio
5. Los que tengan ambas letras 'd' y 'e' en cualquier orden
6. Los que empiezan por 'x' o por 'y'
7. Los que empiezan por 'x' o 'y' y terminan por 'e'
8. Los que tienen tres o m√°s d√≠gitos seguidos
9. Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'

```python
import re
```

```python
accs = ['xkn59438', 'yhdck2', 'eihd39d9', 'chdsye847', 'hedle3455', 'xjhd53e', '45da', 'de37dp']
```

```python
# Los que tengan el n√∫mero 5
# El patr√≥n "5" busca simplemente si aparece el d√≠gito 5 en cualquier parte
print("1. Los que tienen el n√∫mero 5:")
print([a for a in accs if re.search("5", a)])

#Los que tengan la letra 'd' o la 'e'
# El patr√≥n "[de]" significa: contiene al menos una 'd' O una 'e'

print("\n2. Los que tienen 'd' o 'e':")
print([a for a in accs if re.search("[de]", a)])

# Los que tengan las letras 'd' y 'e' en ese orden
# El patr√≥n "d.*e" significa:
# - 'd' seguida de cualquier n√∫mero de caracteres (.*)
# - luego aparece una 'e' m√°s adelante

print("\n3. Los que tienen 'd' y 'e' en ese orden:")
print([a for a in accs if re.search("d.*e", a)])

#Los que tengan las letras 'd' y 'e' en ese orden con una letra entre medio
# El patr√≥n "d.e" significa:
# - 'd' seguida de exactamente un car√°cter cualquiera (.)
# - y luego 'e'

print("\n4. Los que tienen 'd' y 'e' con una letra entre medias:")
print([a for a in accs if re.search("d.e", a)])

# Los que tengan ambas letras 'd' y 'e' en cualquier orden
# El patr√≥n "(d.*e|e.*d)" significa:
# - o bien 'd' aparece antes que 'e'
# - o bien 'e' aparece antes que 'd'

print("\n5. Los que tienen 'd' y 'e' en cualquier orden:")
print([a for a in accs if re.search("(d.*e|e.*d)", a)])

#Los que empiezan por 'x' o por 'y'
# El patr√≥n "^[xy]" significa:
# - ^ ‚Üí principio de la cadena
# - [xy] ‚Üí empieza con 'x' o con 'y'


print("\n6. Los que empiezan por 'x' o 'y':")
print([a for a in accs if re.search("^[xy]", a)])


# Los que empiezan por 'x' o 'y' y terminan por 'e'
# El patr√≥n "^[xy].*e$" significa:
# - ^[xy] ‚Üí empieza por 'x' o 'y'
# - .* ‚Üí cualquier n√∫mero de caracteres
# - e$ ‚Üí termina con 'e'
print("\n7. Los que empiezan por 'x' o 'y' y terminan por 'e':")
print([a for a in accs if re.search("^[xy].*e$", a)])


#  Los que tienen tres o m√°s d√≠gitos seguidos
# El patr√≥n "\d{3,}" significa:
# - \d ‚Üí cualquier d√≠gito
# - {3,} ‚Üí tres o m√°s seguidos

print("\n8. Los que tienen tres o m√°s d√≠gitos seguidos:")
print([a for a in accs if re.search(r"\d{3,}", a)])


# Los que terminan con 'd' seguidos de 'a' o 'r' o 'p'
# El patr√≥n "d[arp]$" significa:
# - 'd' seguida de una de las letras entre corchetes (a, r o p)
# - y termina con eso ($)


print("\n9. Los que terminan con 'da', 'dr' o 'dp':")
print([a for a in accs if re.search("d[arp]$", a)])
```

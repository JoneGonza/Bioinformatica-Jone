##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
dna = "AATGATGAACGAC"
dinucleotidos = ['AA','AT','AG','AC',
                 'TA','TT','TG','TC',
                 'GA','GT','GG','GC',
                 'CA','CT','CG','CC']

conteo = {}

for d in dinucleotidos:
    conteo[d] = dna.count(d)

# Imprimir solo los que tienen count == 2
for d, c in conteo.items():
    if c == 2:
        print(d, "→", c)
```

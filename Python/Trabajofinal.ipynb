#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
# Esto es un diccionario con las enzimas y los patrones que generan, siendo * el punto de corte
```

```python
# Genero y defino una función para la digestión enzimática y que cuando la vaya a usar tendré que darle la secuencia de ADN y las enzimas, su diccionario
def digestion(secuencia, enzimas):
# Primero genero variables para cada función que se me ha pedido
    enzimas_con_corte = {} # aqui se guardan las enzimas que cortan y la posición en la que lo hacen
    enzimas_sin_corte = [] # aqui se guardan las enzimas que no tienen sitio de corte en la secuencia
    todas_las_posiciones = [] # esto me va a dar las posiciones en las que hay corte
    for nombre, patron in enzimas.items(): # recorro el diccionario de enzimas. Es un bucle que en cada iteración, nombre contiene el nombre de la enzima y patron su secuencia de reconocimiento con el asterisco. Sirve para cualquier diccionario de enzimas
        indice_asterisco = patron.index("*") # así descubro dónde está el punto de corte dentro del patrón; es decir la posicion del * en la secuencia de reconocimiento
        sitio = patron.replace("*","") # quito el asterisco para poder buscar el punto real en la secuencia
        posiciones_encontradas = [] # una lista para guardar las posiciones de ese sitio en la secuencia
        inicio_busqueda = 0 # indico que se busque desde el principio de la secuencia
        # Genero un bucle para encontrar todas las posiciones del sitio en la secuencia
        while True:
            pos = secuencia.find(sitio, inicio_busqueda) # busco el sitio y lo llamo pos
            if pos == -1: # si ya no aparece más que pare
                break
            pos_corte = pos + indice_asterisco # el corte enzimático se da en la posición donde inicia el sitio + el índice del asterisco
            posiciones_encontradas.append(pos_corte) # para guardar la posición
            inicio_busqueda = pos + 1 # sigo buscando en adelante
        # Ahora, en caso de que la enzima tenga sitios de corte
        if posiciones_encontradas: #si la enzima encuentra puntos de corte
            enzimas_con_corte[nombre] = posiciones_encontradas # guardo en un diccionario el nombre y las posiciones
            todas_las_posiciones.extend(posiciones_encontradas) # añado esas posciones a la lista global de cortes
        else:
            enzimas_sin_corte.append(nombre) #aquí se guardan las enzimas que no corten
        
    # Voy a eliminar los duplicados por si diferentes enzimas cortaran en el mismo punto
    # Ordeno la lista
    cortes_ordenados = sorted(set(todas_las_posiciones))
    
    #Guardo los fragmentos de ADN que se generan después de cortar con las enzimas
    fragmentos =[]
    inicio = 0 # para que el primer fragmento sea al inicio de la secuencia

    #Para cada punto de corte genero un fragmento
    for corte in cortes_ordenados:
        fragmentos.append(secuencia[inicio:corte])
        inicio = corte # para que el siguiente fragmento empiece desde el corte

    # El último fragmento tiene que ser desde el último corte hasta el final de la secuencia
    fragmentos.append(secuencia[inicio:])

    # Calculo el tamaño de cada fragmento en número de nucleótidos
    tamaños = [len(f) for f in fragmentos]

    # Devuelvo todo lo calculado para luego imprimir
    return {
        "enzimas_con_corte": enzimas_con_corte,
        "enzimas_sin_corte": enzimas_sin_corte,
        "cortes": cortes_ordenados,
        "fragmentos": fragmentos,
        "tamaños": tamaños
    }
    
```

```python
# Ahora voy a usar lo generado y ver los resultados
# Llamo a la función usando mi secuencia base y las enzimas dadas
resultado = digestion(secuencia_entera, enzR)
```

```python
# Voy a sacar las enzimas que sí que cortan
print("=== Enzimas que cortan y posiciones de corte ===") # genero un encabezado, los === son decorativos
for enz, pos in resultado["enzimas_con_corte"].items(): # bucle para recorrer el diccionario de las enzimas con corte y que devuelva pares de clave,valor
    print(f"{enz}: posiciones (0-based) → {pos}") # f string para insertar variables dentro de un texto. enz se reemplaza por el nombre de la enzia y pos por la lista de posiciones en las que corta. Se indica que las posiciones empiezan desde 0
```

```python
# Las enzimas que no cortan
print("\n=== Enzimas que NO cortan ===")
print(resultado["enzimas_sin_corte"])
```

```python
# Hago que salgan los fragmentos resultantes de la disgestión
print("\n=== Fragmentos tras la digestión ===") # genero un salto de línea antes del texto e imprimo el encabezado
for i, fragmento in enumerate(resultado["fragmentos"], start=1): # i es el número del fragmento empezando en 1 por start=1. Con enumerate recorro la lista obteniendo el índice (i) y el valor (fragmento) en mi lista
    print(f"\nFragmento {i}:") # imprimo el número del fragmento con salto de línea
    print(fragmento)
```

```python
# Con len saco el número de elementos de mi lista; cuántos fragmentos han surgido
print(f"\nNúmero total de fragmentos: {len(resultado['fragmentos'])}")
# Tamaño de cada fragmento
for i, fragmento in enumerate(resultado['fragmentos'], start=1):
    print(f"Fragmento {i}: {len(fragmento)} pb")
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
def encontrar_orfs(secuencia, top_k=6, imprimir=True):
    # Preparo la secuencia: quito los espacios y los saltos y hago que se escriba en mayúsculas
    seq = "".join(secuencia.upper().split()) # limpio la entrada
    n = len(seq) # longitud total en nucleótidos

    # Defino cuáles son los codones de STOP y la tabla de complementos. 
    stop_codones = {"TAA", "TAG", "TGA"} # set de codones de parada
    complementos = {"A":"T","T":"A","C":"G","G":"C"} # diccionario de los complementos de los nucleótidos, para sacar el complemento inverso
    codon2aa = {  # diccionario de traducción a aminoácidos, cada codón de ADN a qué aminoácido se traduce (una letra)
        "TTT":"F","TTC":"F","TTA":"L","TTG":"L","CTT":"L","CTC":"L","CTA":"L","CTG":"L",
        "ATT":"I","ATC":"I","ATA":"I","ATG":"M","GTT":"V","GTC":"V","GTA":"V","GTG":"V",
        "TCT":"S","TCC":"S","TCA":"S","TCG":"S","AGT":"S","AGC":"S",
        "CCT":"P","CCC":"P","CCA":"P","CCG":"P",
        "ACT":"T","ACC":"T","ACA":"T","ACG":"T",
        "GCT":"A","GCC":"A","GCA":"A","GCG":"A",
        "TAT":"Y","TAC":"Y",
        "CAT":"H","CAC":"H","CAA":"Q","CAG":"Q",
        "AAT":"N","AAC":"N","AAA":"K","AAG":"K",
        "GAT":"D","GAC":"D","GAA":"E","GAG":"E",
        "TGT":"C","TGC":"C","TGG":"W",
        "CGT":"R","CGC":"R","CGA":"R","CGG":"R","AGA":"R","AGG":"R",
        "GGT":"G","GGC":"G","GGA":"G","GGG":"G"
    }
    arom = {"F","Y","W"} # set de aminoácidos aromáticos

    # Con la función revcomp(s) devuelve la cadena complementaria inversa en secuencia de ADN según las leyes de complementariedad, antiparalelas
    # Defino esa funcion y recibe el parámetro s que es la cadena de ADN
    def revcomp(s):
        return "".join(complementos.get(b, "N") for b in s[::-1]) 
        #s[::-1] invierte el orden de la cadena
        #for b in s[::-1] recorre cada base (b) de la cadena invertida
        #(complementos.get(b, "N") busca el complemento de cada base mediante el diccionario de complementos
        # con .get(b,"N") le digo que me de el valor de b en el diccionario y que si no existe me devuelva N
        #con .join uno las letras, bases, en una cadena

    # Traduzco el ADN a proteína (secuencia de aminoácidos) hasta encontrar un codón STOP
    def traducir(seq_nt): # defino la función traducir que recibe una secuencia de ADN (seq_nt)
        aa = [] #lista vacía donde se irán guardando los aminoácidos traducidos
        for i in range(0, len(seq_nt)-2, 3): # bucle que recorre la secuencia de tres en tres (los codones son tríos de nucleótidos)
            # len(seq_nt)-2 sirve para evitar leer fragmentos incompletos
            cod = seq_nt[i:i+3] #extraigo un codón (3 letras)
            if cod in stop_codones: #si el codón es de stop, se para la traducción
                break
            aa.append(codon2aa.get(cod, "X")) # uso un diccionario genético para traducir el codón a aminoácido, si no existe en el diccionario se traduce X
        return "".join(aa) #une todos los aminoácidos en una cadena

    # Busco ORFs en una hebra 
    def buscar_orfs_en_hebra(seq, hebra_label, coord_transform=None): #seq es la secuencia de ADN a examinar 
        #hebra_label va a indicar si la hebra es positiva (+) o negativa (-)
        #coord_transform en la hebra inversa convierte las coordenadas a la posición real
        encontrados = [] #lista vacía donde se guardan los ORFs encontrados
        L = len(seq) #guardo la longitud de la secuencia
        for frame in range(3): #porque hay 3 posibles modos de lectura: 0= lee desde el primer ncl, 1= lee desde el segundo, 2= lee desde el tercero
            i = frame
            while i <= L - 3: #inicio la lectura desde la posición de ese frame y recorre de tres en tres, me aseguro de no leer codones incompletos (L-3)
                if seq[i:i+3] == "ATG": #busco el codón de inicio, ahí puede empezar un ORF
                    j = i + 3
                    while j <= L - 3: #busco el stop más cerrcano en el mismo marco de lectura
                        if seq[j:j+3] in stop_codones: #llegar a un codón stop indica final de ORF
                            orf_nt = seq[i:j+3] #saco la secuencia completa del ORF desde el codón de inicio hasta el stop, incluido
                            aa = traducir(orf_nt) #traduzco la secuencia a aminoácidos (se traduce sin incluir el stop)
                            # coordenadas en la hebra que se está analizando (coord_transform opcional ajusta a genoma)
                            start_genomic, end_genomic = (i, j+3) if coord_transform is None else coord_transform(i, j+3) # si la hembra es positiva uso(i, j+3), si es negativa uso coord_transform para convertir al sentido correcto
                            encontrados.append({ #dicionario para guardar toda la información del ORF dentro de la lista encontrados
                                "hebra": hebra_label, #positiva o negativa
                                "frame": frame, #marco de lectura
                                "start": start_genomic, #coordenadas del ORF
                                "end": end_genomic, #coordenadas del ORF
                                "len_nt": len(orf_nt), #longitud en nucleótidos
                                "seq_nt": orf_nt, #secuencia de ADN
                                "seq_aa": aa #secuencia de aminoácidos traducida
                            })
                            break #cuando se encuentre STOP, paro la busqueda para ese ORF, luego sigue buscando ATG
                        j += 3
                i += 3
        return encontrados

    # Busco ORFs en la hebra positiva (directas)
    orfs_pos = buscar_orfs_en_hebra(seq, "+", coord_transform=None)

    # Busco ORFs en hebra negativa: busco en la reverse-complement, pero convierto coordenadas
    rc = revcomp(seq)
    # coord_transform para pasar coords de RC a coords en la secuencia original:
    # si en RC tenemos start_rc=i y end_rc=j, en genoma original:
    # start_genomic = n - j, end_genomic = n - i
    def coord_rc_a_genoma(i_rc, j_rc):
        return (n - j_rc, n - i_rc)
    orfs_neg = buscar_orfs_en_hebra(rc, "-", coord_transform=coord_rc_a_genoma)

    # Uno todos los ORFs buscados (directos y reverse)
    todos_orfs = orfs_pos + orfs_neg
    if not todos_orfs: #si no hay ORFs, devuelve un resultado indicativo: frase
        if imprimir:
            print("No se encontraron ORFs.")
        return {"orfs": [], "top": [], "best": None}

    # Ordenar por longitud en nucleótidos (descendente, de mayor a menor)
    orfs_ordenados = sorted(todos_orfs, key=lambda x: x["len_nt"], reverse=True) #ya tenía creada la lista todos_orfs con diccionarios
    #con sorted ordeno la lista de menor a mayor
    #key=lambdax: x["len_nt"] indica el criterio de orden, ordeno usando el valor de longitud de nucleótdios de cada ORF
    #con reverse=True lo invierto para ordenar de mayor a menor
    top_k_orfs = orfs_ordenados[:top_k] #le pido darme los primeros top_k de la lista orfs_ordenados
    best = orfs_ordenados[0]

    # Métricas para el mejor ORF
    proteina = best["seq_aa"]
    aa_len = len(proteina)
    n_arom = sum(1 for a in proteina if a in arom)
    pct_arom = (100.0 * n_arom / aa_len) if aa_len else 0.0
    pct_rel = 100.0 * best["len_nt"] / n

    # Impresión
    if imprimir:
        print(f"Total ORFs encontrados: {len(todos_orfs)}")
        print("\nTop ORFs (hebra, inicio-end, len_nt):")
        for i, o in enumerate(top_k_orfs, 1):
            print(f"{i}. {o['hebra']} | {o['start']}-{o['end']} | {o['len_nt']} nt")
        print("\nMejor ORF (más largo):")
        print(f"Hebra: {best['hebra']}  Frame: {best['frame']}  Pos: {best['start']}-{best['end']}  Len: {best['len_nt']} nt")
        print(f"Proteína: {aa_len} aa  % aromáticos: {pct_arom:.2f}%  % ORF/ADN: {pct_rel:.2f}%")

    return {
        "orfs": todos_orfs,
        "top": top_k_orfs,
        "best": best,
        "protein": proteina,
        "pct_aromaticos": pct_arom,
        "pct_relativo": pct_rel
    }
```

```python
resultado = encontrar_orfs(secuencia_entera)
```
